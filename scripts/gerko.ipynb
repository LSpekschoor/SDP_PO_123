Feature engineering:
-App ID : Zit er nog in zodat je nog kan herleiden welke app het was. Niet voor gebruik in ML model.
-Category: Moet nog omgevormd worden naar bruikbare variabelen
-Rating
-Rating count
-Maximum installs - onze te voorspellen variabele
-Free : Even kijken naar correlatie met prijs
-Price
-Size : Omvormen naar getal
-Minimum Android : omvormen naar getal
-Developer ID: Omvormen naar variabele die het totale aantal apps van deze developer in de appstore aangeeft. Variabele DeveloperID kan dan zelf weg.
-Released: Verschil tussen Released en Scraped Time berekenen -> mogelijk geen goede voorspeller installs door survivorship bias
-Last Updated : Verschil tussen Last Updated en Scraped Time -> een app die vaak wordt geupdate geeft over het algemeen waarschijnlijk meer installs omdat bugs worden verholpen en features toegevoegd
-Content Rating : categorieen. Mogelijk weinig tot geen effect omdat 87% Everyone is.
-Ad supported
-In App Purchases
-Editors Choice
-Scraped time : voor berekeningen, zie hierboven.

Simpele lineaire regressie om te kijken naar performance op basis van de data set zonder feature engineering en alleen dus kijkende naar variabelen die een getal zijn en binaire variabelen.

```python
df_nonan = df.dropna()
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(df_nonan[['Rating', 'Rating Count', 'Free', 'Ad Supported', 'In App Purchases', 'Editors Choice']], df_nonan['Maximum Installs'])
```

```python
r_sq = model.score(df_nonan[['Rating', 'Rating Count', 'Free', 'Ad Supported', 'In App Purchases', 'Editors Choice']], df_nonan['Maximum Installs'])
print(f"coefficient of determination: {r_sq}")
```
